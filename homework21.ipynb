```python
import pandas as pd
```

```python
df = pd.read_csv('vgsales.csv')
df
```

```python
df = df.dropna()
df.describe()
```

Среднее по всем играм - 70, так что ставим условие выше этого среднего   
H0: средняя удовлетворенность критиками спортивными играми <= 70  
H1: средняя удовлетворенность критиками спортивными играми более 70

```python
df_sport = df[df['Genre'] == 'Sports']
df_sport
```

```python
df_sport.describe()
```

```python
from scipy import stats as st
alpha = 0.05
result = st.ttest_1samp(df_sport['Critic_Score'], 70, alternative='greater')

print(df_sport['Critic_Score'].mean())
print(result)

if result.pvalue < alpha: 
    print('Отвергаем нулевую гипотезу, среднее больше 70 (игры нравятся)')
else:
    print('Не отвергаем нулевую гипотезу')
```

Получается, что в среднем критики ставят высокие оценки спортивным играм

H0: оценки критиков игры на PC не отличаются от оценок игр на PS4   
H1: оценки критиков игры на PC отличаются от оценок игр на PS4

```python
df_pc = df[df['Platform'] == 'PC'][['Critic_Score']]
df_pc
```

```python
df_ps = df[df['Platform'] == 'PS4'][['Critic_Score']]
df_ps
```

```python
st.ttest_ind(df_pc, df_ps, equal_var=False)
```

pvalue меньше 0.05 - различия не существенны

H0: оценки критиков для стрелялок не отличаются от оценок стратегий   
H1: оценки критиков для стрелялок отличаются от оценок стратегий

```python
df = pd.read_csv('spam.csv')
df
```

```python
df['Message'] = df['Message'].str.lower()
df
```

```python
import re
df['Message'] = df['Message'].apply(lambda x: re.sub('[\W_]+', ' ', x))
df
```

```python
import nltk
from nltk.corpus import stopwords
stopwords_set = set(stopwords.words('english'))
texts_splitted = [text.split() for text in df['Message']]
without_sw = [[word for word in text if word not in stopwords_set] for text in texts_splitted]
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
lemmatized_text = [[wordnet_lemmatizer.lemmatize(word) for word in text] for text in without_sw]
```

```python
df['Message'] = [' '.join(row) for row in lemmatized_text]
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(df.Message)
names = tfidf.get_feature_names()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X = tfidf_matrix 
y = df['Category']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

т.е. хорошее сообщение ошибочно 3 раза попало в спам, а спам 66 раз попал в хорошие
